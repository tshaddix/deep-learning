## Overfitting and Underfitting

- Overfitting (high variance)
- Underfitting (high bias)

## Model Complexity & Early Stopping



You can do the gradient descent until the testing error stops decreasing and starts increasing. That's the epoch when we stop. This algorithm is called early stopping.

## Regularization

Punishing large coefficients by some factor in order to account for issues introduced when using large coefficients.





## Dropout

Randomly turn off some nodes to force other nodes to contribute. Turn off different nodes during each epoch. Given by some probability that each node will get dropped.

## Learning Rate



General rule of thumb - if your model is not performing, decrease the learning rate.
