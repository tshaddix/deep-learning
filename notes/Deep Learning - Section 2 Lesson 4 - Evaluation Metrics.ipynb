Accuracy does not always work as a proper measure of a model. Consider a model that has 200,000 positive examples and 200 negative examples. In this case, even if the model classified 100% of the data as positive, you would still get an incredibly high accuracy score.

## Precision and Recall

High Recall - False positives ok, False negatives not ok
High Precision - False positives not ok, false negatives ok

Precision - out of all the points predicted to be positive, how many were actually positive?

Precision = true positives / (true positive + false positives)

Recall - How many of the positive points did I manage to catch?

Recall = true positives / (true positives + false negatives)
